```python
import pandas as pd
import pickle
import re
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

tring Cleansing

```python
def data_processing(text):
    text = text.lower()
    text = re.sub(r"https\S+|www\S+https\S+", '',text, flags=re.MULTILINE)
    text = re.sub(r'\@w+|\#','',text)
    text = re.sub(r'[^\w\s]','',text)
    text_tokens = word_tokenize(text)
    filtered_text = [w for w in text_tokens if not w in stop_words]
    return " ".join(filtered_text)
```

emmatization

```python
def lemmatizing_frame(data):
    lempred = WordNetLemmatizer()
    l = len(data)
    for i in range(0,l,1):
        txt = []
        list = data[i].split(" ")
    
    
        for w in list:
            a = lempred.lemmatize(w)
            txt.append(a)
            textprocessed = ' '.join(map(str,txt))
        data[i] = textprocessed
```

```python
def finalprediction(data):
    predict_ip = "trump administration failed to deliver vaccines to people"
    predf = pd.DataFrame()
    predf['text'] = [predict_ip]
    predf.text = predf['text'].apply(data_processing)
    lemmatizing_frame(predf['text'])
    predf = predf.drop_duplicates('text')
    Xpred = predf['text']
    
    filesvc = 'finalized_model_SVC.sav'
    loaded_model = pickle.load(open(filesvc,'rb'))
    vector = "vectoriser.sav"
    load_vector = pickle.load(open(vector,'rb'))
    Xpred = load_vector.transform(Xpred)
    svc_pred = loaded_model.predict(Xpred)
    return str(svc_pred[0])
```

```python
print(finalprediction("trump administration failed to deliver vaccines to people"))
```
